```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
train = pd.read_csv('../data/interim/train.csv')
test = pd.read_csv('../data/raw/test.csv')
```

```python
test['datetime'] = pd.to_datetime(test['datetime'])
test['hour'] = test['datetime'].apply(lambda x:x.hour)
test['month'] = test['datetime'].apply(lambda x:x.month)
test['day_of_week'] = test['datetime'].apply(lambda x:x.dayofweek)
```

```python
y_train = train['count']
y_train.to_csv('../data/processed/y_train.csv', index=False)
x_train = train[['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','month','day_of_week']]
test = test[['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','month','day_of_week']]
```

```python
x_train['windspeed'].replace(0, np.nan, inplace=True)
x_train['windspeed'] = x_train['windspeed'].fillna(x_train.groupby('weather')['season'].transform('mean'))
```

```python
x_train = pd.get_dummies(data=x_train, columns=['season', 'holiday', 'workingday', 'weather'])
test = pd.get_dummies(data=test, columns=['season', 'holiday', 'workingday', 'weather'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler = scaler.fit(x_train[['temp', 'atemp', 'humidity', 'windspeed']])
x_train[['temp', 'atemp', 'humidity', 'windspeed']] = scaler.transform(x_train[['temp', 'atemp', 'humidity', 'windspeed']])
test[['temp', 'atemp', 'humidity', 'windspeed']] = scaler.transform(test[['temp', 'atemp', 'humidity', 'windspeed']])
```

```python
x_train.to_csv('../data/processed/x_train.csv', index=False)
test.to_csv('../data/processed/test.csv', index=False)
```
